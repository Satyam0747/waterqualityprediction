```python
from tkinter import *
import joblib
def show_entry_fields():
    p1=float(e1.get())
    p2=float(e2.get())
    p3=float(e3.get())
    p4=float(e4.get())
    p5=float(e5.get())
    p6=float(e6.get())
    p7=float(e7.get())
    p8=float(e8.get())
    p9=float(e9.get())
    model = joblib.load('quality_model')
    result=model.predict([[p1,p2,p3,p4,p5,p6,p7,p8,p9]])
    
    if result == 0:
        Label(master, text="Water is not drinkable").grid(row=31)
    else:
        Label(master, text="Water is drinkable").grid(row=31)
    
    
master = Tk()
master.title("Water Quality Prediction")


label = Label(master, text = "Water Quality Prediction"
                          , bg = "black", fg = "white"). \
                               grid(row=0,columnspan=2)


Label(master, text="pH").grid(row=1)
Label(master, text="Hardness").grid(row=2)
Label(master, text="Solids").grid(row=3)
Label(master, text="Chloramines").grid(row=4)
Label(master, text="Sulfate").grid(row=5)
Label(master, text="Conductivity").grid(row=6)
Label(master, text="Organic_carbon").grid(row=7)
Label(master, text="Trihalomethanes").grid(row=8)
Label(master, text="Turbidity").grid(row=9)




e1 = Entry(master)
e2 = Entry(master)
e3 = Entry(master)
e4 = Entry(master)
e5 = Entry(master)
e6 = Entry(master)
e7 = Entry(master)
e8 = Entry(master)
e9 = Entry(master)

e1.grid(row=1, column=1)
e2.grid(row=2, column=1)
e3.grid(row=3, column=1)
e4.grid(row=4, column=1)
e5.grid(row=5, column=1)
e6.grid(row=6, column=1)
e7.grid(row=7, column=1)
e8.grid(row=8, column=1)
e9.grid(row=9, column=1)



Button(master, text='Predict', command=show_entry_fields).grid()

mainloop()
```
